```python
driver=webdriver.Chrome(executable_path="C:\webdriver\chromedriver.exe")
driver.get("https://www.zhihu.com/question/306275416")
driver.set_window_size(400,800)
time.sleep(10)
# prefs = {"profile.managed_default_content_settings.images":2}
# chrome_options=Options()
# chrome_options.add_experimental_option("prefs",prefs)
            
def tuo():
    js="var q=document.documentElement.scrollTop=100000000000"
    driver.execute_script(js)
for i in range(50):
    time.sleep(0.2)
    tuo()
    tuo()
    tuo()
    tuo()
    tuo()
    tuo()
    tuo()
    tuo()
    tuo()
    tuo()
    tuo()
    tuo()
    tuo()
    tuo()
    tuo()
    tuo()
    tuo()
    tuo()
    print("第{}次循环".format(i))

s_selector=Selector(text=driver.page_source)
divs=s_selector.xpath("//div[@class='RichContent-inner']")
dis=s_selector.xpath("//div[@class='ztext AuthorInfo-badgeText']")
# print(divs)
Allcontent=[]
# Allcontent=pd.DataFrame(Allcontent)
for  div, di in zip(divs,dis):
    
    content=div.xpath(".//span//p/text()").extract()
    content = ''.join(content)
    qm=di.xpath("./text()").extract()
    qm = ''.join(qm)
    
#     aa=pd.DataFrame(zip([qm,content]))
    aa=[qm,content]

    Allcontent.append(aa)
Allcontent=pd.DataFrame(Allcontent)
Allcontent.to_csv("C:/Users/dgc/Desktop/Allcontent.csv",encoding='gb18030')
```
