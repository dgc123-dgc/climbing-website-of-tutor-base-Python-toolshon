```python
import urllib 
import urllib.request
import re
import pandas as pd

def getpage(path):
    data=urllib.request.urlopen(path).read().decode("utf-8")
    return data
# def getinfo():

def change(data):
    ss=[]
    for i in data:
        
        i=i.strip()
        ss.append(i)
    return ss
def remo(a):
    for i in a:
        if i=="":
            a.remove(i)
    return a
def sp(data):
    data=data.split('[')
    return data

def combine(data):
    starturl="http://www.ygjj.com/"
    for i in range(len(data)):
        data[i]=starturl+data[i]
    return data

def getlink(url):
    data = getpage(url)
    from scrapy.selector import Selector
    selector=Selector(text=data)
    suburl=selector.xpath("//div [@class='dd_title']//a/@href").extract()
    suburl=combine(suburl)
    starturl="http://www.ygjj.com/"
    print(suburl)
    return suburl
def remokg(data):
    for i in range(len(data)):
        data[i]=data[i].strip()
    return data
    
def genlist():
    s="s."*15
    s=s.split(".")
    s.pop()
    for i in range(15):
        s[i]=s[i]+str(i+1)
    return s
def judge(data):
    for i in data:
        if i in ["发布中" , "已关闭"]:
            b=i
    return b 
        
    
    
def getdata(url):
    data = getpage(url)
    from scrapy.selector import Selector
    selector=Selector(text=data)
    data=genlist()
    s1=selector.xpath("//div [@class='ch2']//a[position()=1]/text()").extract()[0]#学科
    s2=selector.xpath("//div [@class='ch2']//a[position()=1]/text()").extract()[1]#学员年级
    s3=selector.xpath("//div [@class='row']/div[@class='ch4']//p[position()=1]/text()").extract()#学员性别
    s4=selector.xpath("//div [@class='row']/div[@class='ch4']//p[position()=2]/text()").extract()#学员性别
    s5=selector.xpath("//div [@class='row2']//td/text()").extract()#薪酬标准
    s6=selector.xpath("//div [@class='dd_detail_lstit']//span/text()").extract()
    s5=change(s5)
    s4=change(s4)
    s3=remokg(s3)
    s4=remokg(s4)
    if len(s5)==15:
        s5.pop()
    elif len(s5)>=16:
        s5=s5[:14]
    if len(s6)>1:
        s6=judge(s6)
    print(s6)

    S=[s1]+[s2]+s3+s4+s5+[s6]
    column=['学科','科目','学员性别','学员年龄','薪酬水平','乘车补贴','学员状态','就读学校','教授时间','blank','居住区域','交通线路','blank','性别要求','需求人数','资格要求1','资格要求2','教学方式','状态']
    S=pd.DataFrame(S)
    S=S.T
    S.columns=column
    print(S.head())
    return S
def getdatatwo(link):
    
    links=getlink(link)
    ss=[]
    for i in range(len(links)):
        print(links[i])
        ss.append(getdata(links[i]))
    #     print(getdata(links[i]))
    print(ss)
    column=['学科','科目','学员性别','学员年龄','薪酬水平','乘车补贴','学员状态','就读学校','教授时间','blank','居住区域','交通线路','blank','性别要求','需求人数','资格要求1','资格要求2','教学方式','状态']

    ss=pd.concat(ss,axis=0,ignore_index=True)
    return ss
```
